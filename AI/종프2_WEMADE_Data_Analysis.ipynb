```python
x_xp = len(df.loc[(df['DI1_dg'] == 0) & (df['sm_present'] == 0)])
x_op = len(df.loc[(df['DI1_dg'] == 0) & (df['sm_present'] == 1)])
o_xp = len(df.loc[(df['DI1_dg'] == 1) & (df['sm_present'] == 0)])
o_op = len(df.loc[(df['DI1_dg'] == 1) & (df['sm_present'] == 1)])
print('고혈압 X, 흡연 X :',x_xp)
print('고혈압 X, 흡연 O :',x_op)
print('고혈압 X의 흡연자 비율 :', x_op / (x_xp + x_op))
print('고혈압 O, 흡연 X :',o_xp)
print('고혈압 O, 흡연 O :',o_op)
print('고혈압 O의 흡연자 비율 :', o_op / (o_xp + o_op))
```

```python
x_xw = len(df.loc[(df['DI1_dg'] == 0) & (df['pa_walk'] == 0)])
x_ow = len(df.loc[(df['DI1_dg'] == 0) & (df['pa_walk'] == 1)])
o_xw = len(df.loc[(df['DI1_dg'] == 1) & (df['pa_walk'] == 0)])
o_ow = len(df.loc[(df['DI1_dg'] == 1) & (df['pa_walk'] == 1)])
print('고혈압 X, 유산소운동 X :',x_xw)
print('고혈압 X, 유산소운동 O :',x_ow)
print('고혈압 X의 유산소운동 비율 :', x_ow / (x_xw + x_ow))
print('고혈압 O, 유산소운동 X :',o_xw)
print('고혈압 O, 유산소운동 O :',o_ow)
print('고혈압 O의 유산소운동 비율 :', o_ow / (o_xw + o_ow))
```
