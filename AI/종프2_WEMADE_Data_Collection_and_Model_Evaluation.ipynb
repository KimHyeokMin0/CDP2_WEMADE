```python
selected_all = pd.concat([selected05, selected07, selected08, selected09, selected10, selected11, selected12, selected13, selected14,
                          selected15, selected16, selected17, selected18, selected19, selected20, selected21, selected22], axis=0, ignore_index=True)

idx = selected_all[selected_all['DI1_dg'] == 9].index
selected_all = selected_all.drop(idx)
selected_all = selected_all.sort_index(ascending=False)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(selected_all[['sex', 'age', 'HE_sbp', 'HE_dbp', 'HE_PLS', 'HE_BMI', 'sm_present', 'pa_walk', 'total_sleep']],
                                                    selected_all['DI1_dg'], test_size = 0.2, stratify = selected_all['DI1_dg'])

x_train = x_train.astype('float')
x_test = x_test.astype('float')
y_train = y_train.astype('int')
y_test = y_test.astype('int')

print(x_train)
print(y_train)
print('-' * 20)
print(x_test)
print(y_test)
```

```python
def normalize_column(column):
    return (column - column.min()) / (column.max() - column.min())

x_train = x_train.apply(normalize_column)
x_test = x_test.apply(normalize_column)
```

```python
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

mlp = Sequential([
    keras.layers.Input(shape=(9,)),
    keras.layers.Dense(7, activation='tanh'),
    keras.layers.Dense(7, activation='tanh'),
    keras.layers.Dense(7, activation='tanh'),
    keras.layers.Dense(1, activation='sigmoid'),
])

mlp.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
history = mlp.fit(x_train, y_train, batch_size=128, epochs=50, validation_data=(x_test, y_test))

res = mlp.evaluate(x_test, y_test)
print('정확률=', res[1])
```

# 모델 평가 (혼동 행렬 + ROC 커브, AUC 면적)

```python
# confusion_matrix 써보기
pred = mlp.predict(x_test)
pred = pred // 0.5
pred = pred.reshape(-1).astype(int)
y_test_numpy = y_test.to_numpy()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test_numpy, pred)
display(cm)

cm_df = pd.DataFrame(cm, index=['고혈압X','고혈압O'], columns=['고혈압 예측 X','고혈압 예측 O'])
display(cm_df)

from sklearn.metrics import classification_report
print(classification_report(y_test_numpy, pred))
```

```python
# ROC 커브
from sklearn.metrics import roc_curve

y_score = mlp.predict(x_test)
fpr, tpr, thresholds = roc_curve(y_test, y_score)
```

```python
import matplotlib.pyplot as plt

plt.plot(fpr, tpr, color = 'red', label = 'ROC curve')
plt.plot([0, 1], [0, 1], color = 'blue', label = 'y = x') # y = x 직선 표시

plt.xlabel('FP Rate')
plt.ylabel('TP Rate')

plt.legend() # 그래프 라벨 표시

plt.show()
```

```python
# AUC 면적
from sklearn.metrics import roc_auc_score

print(roc_auc_score(y_test, y_score))
```
